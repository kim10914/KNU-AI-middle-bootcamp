```python
from openai import OpenAI
from dotenv import load_dotenv
import os

load_dotenv()
# Whisper는 OpenAI 정식 키가 필요합니다. (OpenRouter sk-or-v... 키는 음성 인식 미지원)
api_key = os.getenv("OPENAI_DIRECT_KEY")
client = OpenAI(api_key=api_key)
```

```python
audio_file_path = "./audio/lsy_audio_2023_58s.mp3"

with open(audio_file_path, "rb") as audio_file:
    transcription = client.audio.transcriptions.create(
        model="whisper-1",
        file=audio_file
    )

transcription
```

```python
# 위스퍼 API로 한국어 음성 파일을 영어로 바로 번역하기
# transcriptions(받아쓰기) 대신 translations를 쓰면 영어로 번역해 반환한다.
with open(audio_file_path, "rb") as audio_file:
    translation = client.audio.translations.create(
        model="whisper-1",
        file=audio_file,
    )

translation
```
