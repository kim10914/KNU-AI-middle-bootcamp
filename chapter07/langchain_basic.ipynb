# 랭체인으로 챗봇 만들기 (가장 기본 형태)

- 기존에는 openai SDK 로 직접 API 를 호출했지만, 랭체인을 쓰면 `ChatOpenAI` 같은 래퍼로 언어 모델을 선언만 해두고 호출한다.
- 나중에 다른 언어 모델(클로드·제미나이 등)로 바꿔도 '모델 선언 부분'만 교체하면 되므로 특정 모델에 종속되지 않는다.
- 이 예제는 단발성 호출이라 이전 대화를 기억하지 못한다.

```python
# 랭체인 설치 (한 번만 실행)
%pip install langchain langchain-openai
```

```python
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
import os

load_dotenv()  # .env 파일의 환경변수 로드

# 언어 모델 선언 (OpenRouter 경유)
model = ChatOpenAI(
    model="openai/gpt-4o-mini",
    base_url="https://openrouter.ai/api/v1",  # OpenRouter 를 쓰므로 필수!
    api_key=os.getenv('OPENAI_API_KEY'),       # OpenRouter 전용 키
)
```

```python
from langchain_core.messages import HumanMessage

# 사람(HumanMessage) 메시지 하나를 보내고 답변(AIMessage)을 받는다.
model.invoke([HumanMessage(content="안녕? 나는 이병일 이야.")])
```

```python
# 이전 대화를 기록하지 않으므로 모델은 내 이름을 기억하지 못한다.
model.invoke([HumanMessage(content="내 이름이 뭐지?")])
```
