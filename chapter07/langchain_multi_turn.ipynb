# 랭체인으로 멀티턴 대화하기

기존 메시지 리스트에 추가하는 방식으로 구현한 멀티턴.

- 언어 모델은 '상태'가 없어서 매번 이전 대화를 함께 보내줘야 기억한다.
- 가장 단순한 방법은 messages 리스트에 사람/AI 메시지를 계속 append 하는 것.
    - HumanMessage  : 사용자 발화
    - AIMessage     : 모델 응답
    - SystemMessage : 챗봇의 역할·성격을 정해주는 초기 지시문
- 셀을 실행하면 입력창이 뜬다. `exit` 를 입력하면 종료.

```python
from dotenv import load_dotenv
import os
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage, AIMessage, SystemMessage

load_dotenv()

llm = ChatOpenAI(
    model="openai/gpt-4o-mini",
    base_url="https://openrouter.ai/api/v1",  # 필수!
    api_key=os.getenv('OPENAI_API_KEY'),       # OpenRouter 전용 키
)
```

```python
# 대화 기록. 첫 줄에 시스템 메시지를 넣어 챗봇의 역할을 정해준다.
messages = [
    SystemMessage("너는 사용자를 도와주는 상담사야."),  # 초기 시스템 메시지
]

while True:
    user_input = input("사용자: ")  # 사용자 입력 받기

    if user_input == "exit":  # 사용자가 대화를 종료하려는지 확인
        break

    # 사용자 메시지를 대화 기록에 추가
    messages.append(HumanMessage(user_input))

    # 지금까지의 대화 기록 전체를 기반으로 AI 응답 가져오기
    ai_response = llm.invoke(messages)

    # AI 응답을 대화 기록에 추가 (다음 턴에서 맥락으로 사용됨)
    messages.append(ai_response)

    print("사용자: " + user_input)
    print("AI: " + ai_response.content)  # AI 응답 출력
```
