# 유튜브 영상 요약하기

유튜브에서 영상을 검색하고, 자막을 불러와 요약한다.

## 1. 유튜브 영상 검색

```python
from youtube_search import YoutubeSearch

videos = YoutubeSearch("파뿌리24", max_results=5).to_dict()
videos
```

## 2. 영상 URL 만들기

```python
video_url = 'http://youtube.com' + videos[3]['url_suffix']
video_url
```

## 3. YoutubeLoader 로 자막(내용) 불러오기

```python
from langchain_community.document_loaders import YoutubeLoader

loader = YoutubeLoader.from_youtube_url(
    video_url,
    language=['ko', 'en'],  # 자막 언어
)

loader.load()
```

## 4. 검색된 모든 영상의 내용 불러오기

```python
for v in videos:
    # url_suffix 를 이용하여 video_url 을 만든다.
    v['video_url'] = 'http://youtube.com' + v['url_suffix']

    # YoutubeLoader 를 이용하여 비디오를 로드한다.
    loader = YoutubeLoader.from_youtube_url(
        v['video_url'],
        language=['ko', 'en'],
    )

    v['content'] = loader.load()

videos
```

## 5. 60분 이하 영상만 남기기

```python
print('총 영상 수:', len(videos))
# 영상 길이가 60분 이하인 영상만 남긴다.
videos = [v for v in videos if len(v['duration'].split(':')) < 3]
print('60분 이하 영상 수:', len(videos))
```

## 6. 언어 모델 준비

```python
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
import os

load_dotenv()

model = ChatOpenAI(
    model="openai/gpt-4o-mini",
    base_url="https://openrouter.ai/api/v1",
    api_key=os.getenv("OPENAI_API_KEY"),
)
model.invoke('안녕?')  # 언어 모델이 잘 설정되었는지 테스트
```

## 7. 영상 요약 체인 만들기 (create_stuff_documents_chain)

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_classic.chains.combine_documents import create_stuff_documents_chain

# 동영상 요약 프롬프트 작성
prompt = ChatPromptTemplate.from_messages(
    [("system", "다음 영상에 대한 요약을 한국어로 만들어줘 :\n\n{context}")]
)

chain = create_stuff_documents_chain(model, prompt)

result = chain.invoke({"context": videos[0]['content']})
result
```

## 8. 모든 영상에 대해 요약 생성

```python
from tqdm import tqdm  # tqdm 은 진행 상황을 보여주는 라이브러리

for v in tqdm(videos):
    v['summary'] = chain.invoke({"context": v['content']})

videos
```

## 9. 요약 결과만 보기

```python
for v in videos:
    print('제목:', v['title'])
    print('요약:', v.get('summary'))
    print('-' * 50)
```
